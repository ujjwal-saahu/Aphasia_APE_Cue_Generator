```python
import sys
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT / "src"))
```

```python
from config import FINAL_GENERATED_CUES_BY_CATEGORY_PATH
```

```python
final_cues_df = pd.read_csv(FINAL_GENERATED_CUES_BY_CATEGORY_PATH)

print("FINAL CUES SHAPE:", final_cues_df.shape)
display(final_cues_df.head(20))
```

```python
# keep only columns needed for final master cue table
final_cues_master_df = final_cues_df[[
    "target_word",
    "subcategory",
    "best_prompt",
    "reference_best_human_cue",
    "final_generated_cue",
    "proxy_score",
]].copy()

final_cues_master_df = final_cues_master_df.sort_values(
    ["subcategory", "proxy_score"],
    ascending=[True, False]
).reset_index(drop=True)

print("FINAL MASTER SHAPE:", final_cues_master_df.shape)
display(final_cues_master_df.head(20))
```

```python
# therapist review sheet
therapist_review_df = final_cues_master_df.copy()
therapist_review_df["manual_label"] = ""
therapist_review_df["manual_notes"] = ""

display(therapist_review_df.head(20))
```

```python
def build_image_prompt(target_word: str, subcategory: str) -> str:
    target_word = str(target_word).strip()
    subcategory = str(subcategory).strip()

    if subcategory == "水果":
        return f"a realistic {target_word}, single object, plain white background, high clarity"
    elif subcategory == "蔬菜":
        return f"a realistic {target_word}, single object, plain white background, high clarity"
    elif subcategory in ["動物"]:
        return f"a realistic {target_word}, single subject, clean background, high clarity"
    elif subcategory in ["交通工具"]:
        return f"a realistic {target_word}, clear full view, simple background, high clarity"
    elif subcategory in ["交通場所", "娛樂場所", "學校"]:
        return f"a realistic scene of {target_word}, simple and clear environment, not crowded, high clarity"
    elif subcategory in ["生活用品", "電器用品", "家具", "廚房用品", "文具", "服飾配件"]:
        return f"a realistic {target_word}, single main object, plain background, high clarity"
    elif subcategory in ["動作", "休閒娛樂"]:
        return f"a realistic illustration of the action or activity {target_word}, simple scene, high clarity"
    elif subcategory in ["喝", "食物"]:
        return f"a realistic {target_word}, clear food or drink presentation, plain background, high clarity"
    else:
        return f"a realistic {target_word}, single clear subject, plain background, high clarity"
```

```python
image_generation_df = final_cues_master_df.copy()
image_generation_df["image_prompt"] = image_generation_df.apply(
    lambda row: build_image_prompt(row["target_word"], row["subcategory"]),
    axis=1
)

display(image_generation_df.head(20))
```

```python
FINAL_DIR = PROJECT_ROOT / "data" / "final"

final_cues_master_path = FINAL_DIR / "final_cues_master.csv"
therapist_review_path = FINAL_DIR / "final_cues_for_therapist_review.csv"
image_generation_path = FINAL_DIR / "final_cues_for_image_generation.csv"

final_cues_master_df.to_csv(final_cues_master_path, index=False, encoding="utf-8-sig")
therapist_review_df.to_csv(therapist_review_path, index=False, encoding="utf-8-sig")
image_generation_df.to_csv(image_generation_path, index=False, encoding="utf-8-sig")

print("Saved:", final_cues_master_path)
print("Saved:", therapist_review_path)
print("Saved:", image_generation_path)
```
