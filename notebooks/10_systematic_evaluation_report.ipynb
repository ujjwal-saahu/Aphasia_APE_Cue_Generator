```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT / "src"))
```

```python
from config import (
    CATEGORY_ROUND0_SUMMARY_PATH,
    CATEGORY_APE_ALL_SUMMARY_PATH,
    BEST_PROMPT_BY_CATEGORY_PATH,
    FINAL_GENERATED_CUES_BY_CATEGORY_PATH,
)
```

```python
round0_summary = pd.read_csv(CATEGORY_ROUND0_SUMMARY_PATH)

try:
    ape_summary = pd.read_csv(CATEGORY_APE_ALL_SUMMARY_PATH)
except Exception:
    ape_summary = pd.DataFrame()

best_prompt_by_category_df = pd.read_csv(BEST_PROMPT_BY_CATEGORY_PATH)
final_cues_df = pd.read_csv(FINAL_GENERATED_CUES_BY_CATEGORY_PATH)

print("round0_summary:", round0_summary.shape)
print("ape_summary:", ape_summary.shape if not ape_summary.empty else "(0, 0)")
print("best_prompt_by_category_df:", best_prompt_by_category_df.shape)
print("final_cues_df:", final_cues_df.shape)
```

```python
if ape_summary.empty:
    all_prompt_summary_df = round0_summary.copy()
else:
    all_prompt_summary_df = pd.concat([round0_summary, ape_summary], ignore_index=True)

print("all_prompt_summary_df:", all_prompt_summary_df.shape)
display(all_prompt_summary_df.head(20))
```

```python
# prompt comparison table
prompt_eval_by_category_df = all_prompt_summary_df.copy()

if "selection_score" not in prompt_eval_by_category_df.columns:
    prompt_eval_by_category_df["selection_score"] = np.nan

prompt_eval_by_category_df = prompt_eval_by_category_df.sort_values(
    ["subcategory", "mean_proxy_score"],
    ascending=[True, False]
).reset_index(drop=True)

display(prompt_eval_by_category_df.head(30))
```

```python
# category-level final performance summary
category_performance_df = (
    final_cues_df.groupby("subcategory", as_index=False)
    .agg(
        n_words=("target_word", "count"),
        avg_proxy_score=("proxy_score", "mean"),
        avg_similarity=("semantic_similarity", "mean"),
        avg_brevity=("brevity_score", "mean"),
        leak_rate=("target_leak_penalty", "mean"),
    )
    .sort_values("avg_proxy_score", ascending=False)
    .reset_index(drop=True)
)

# empty cue rate
empty_rate_df = (
    final_cues_df.assign(
        is_empty=final_cues_df["final_generated_cue"].fillna("").astype(str).str.strip().eq("")
    )
    .groupby("subcategory", as_index=False)
    .agg(empty_rate=("is_empty", "mean"))
)

category_performance_df = category_performance_df.merge(
    empty_rate_df,
    on="subcategory",
    how="left"
)

display(category_performance_df)
```

```python
# Best prompts table for professor viewing
best_prompt_report_df = best_prompt_by_category_df.copy()

keep_cols = [c for c in [
    "subcategory",
    "prompt_variant",
    "mean_proxy_score",
    "selection_score",
    "prompt_text"
] if c in best_prompt_report_df.columns]

best_prompt_report_df = best_prompt_report_df[keep_cols].copy()
display(best_prompt_report_df)
```

```python
# top cues by category
top_cues_by_category = (
    final_cues_df
    .sort_values(["subcategory", "proxy_score"], ascending=[True, False])
    .groupby("subcategory", as_index=False)
    .head(5)
    .reset_index(drop=True)
)

display(top_cues_by_category[[
    "subcategory",
    "target_word",
    "reference_best_human_cue",
    "final_generated_cue",
    "proxy_score"
]])
```

```python
# worst cues by category
bottom_cues_by_category = (
    final_cues_df
    .sort_values(["subcategory", "proxy_score"], ascending=[True, True])
    .groupby("subcategory", as_index=False)
    .head(5)
    .reset_index(drop=True)
)

display(bottom_cues_by_category[[
    "subcategory",
    "target_word",
    "reference_best_human_cue",
    "final_generated_cue",
    "proxy_score"
]])
```

```python
# manual review sheet
manual_review_sheet_df = final_cues_df.copy()

manual_review_sheet_df["manual_label"] = ""
manual_review_sheet_df["notes"] = ""

manual_review_sheet_df = manual_review_sheet_df[[
    "subcategory",
    "target_word",
    "reference_best_human_cue",
    "final_generated_cue",
    "proxy_score",
    "manual_label",
    "notes",
]]

display(manual_review_sheet_df.head(20))
```

```python
# APE vs final best prompt summary
# This helps explain whether prompt selection improved over base prompts
base_prompt_summary_df = all_prompt_summary_df.copy()

if "prompt_variant" in base_prompt_summary_df.columns:
    base_only_df = base_prompt_summary_df[
        base_prompt_summary_df["prompt_variant"] == "v1_meta_base"
    ].copy()
else:
    base_only_df = pd.DataFrame()

if not base_only_df.empty:
    base_only_df = (
        base_only_df.sort_values(["subcategory", "mean_proxy_score"], ascending=[True, False])
        .groupby("subcategory", as_index=False)
        .head(1)
        .rename(columns={"mean_proxy_score": "base_prompt_score"})
    )

    compare_df = best_prompt_by_category_df.merge(
        base_only_df[["subcategory", "base_prompt_score"]],
        on="subcategory",
        how="left"
    )

    compare_df["improvement_vs_base"] = (
        compare_df["mean_proxy_score"] - compare_df["base_prompt_score"]
    )

    display(compare_df[[
        "subcategory",
        "prompt_variant",
        "base_prompt_score",
        "mean_proxy_score",
        "improvement_vs_base"
    ]])
else:
    print("No v1_meta_base prompt_variant found, skipping base comparison.")
```

```python
# Save all report files
REPORT_DIR = PROJECT_ROOT / "data" / "final"

prompt_eval_by_category_path = REPORT_DIR / "prompt_eval_by_category.csv"
category_performance_path = REPORT_DIR / "category_performance_summary.csv"
top_cues_path = REPORT_DIR / "top_cues_by_category.csv"
bottom_cues_path = REPORT_DIR / "bottom_cues_by_category.csv"
manual_review_path = REPORT_DIR / "manual_review_sheet.csv"

prompt_eval_by_category_df.to_csv(prompt_eval_by_category_path, index=False, encoding="utf-8-sig")
category_performance_df.to_csv(category_performance_path, index=False, encoding="utf-8-sig")
top_cues_by_category.to_csv(top_cues_path, index=False, encoding="utf-8-sig")
bottom_cues_by_category.to_csv(bottom_cues_path, index=False, encoding="utf-8-sig")
manual_review_sheet_df.to_csv(manual_review_path, index=False, encoding="utf-8-sig")

print("Saved:", prompt_eval_by_category_path)
print("Saved:", category_performance_path)
print("Saved:", top_cues_path)
print("Saved:", bottom_cues_path)
print("Saved:", manual_review_path)
```

```python
# professor-style readable summary
summary_txt_path = REPORT_DIR / "systematic_evaluation_summary.txt"

with open(summary_txt_path, "w", encoding="utf-8") as f:
    f.write("SYSTEMATIC EVALUATION REPORT\n")
    f.write("=" * 100 + "\n\n")

    f.write("1. CATEGORY PERFORMANCE SUMMARY\n")
    f.write("-" * 100 + "\n")
    for _, row in category_performance_df.iterrows():
        f.write(f"Category: {row['subcategory']}\n")
        f.write(f"  n_words: {row['n_words']}\n")
        f.write(f"  avg_proxy_score: {row['avg_proxy_score']:.4f}\n")
        f.write(f"  avg_similarity: {row['avg_similarity']:.4f}\n")
        f.write(f"  avg_brevity: {row['avg_brevity']:.4f}\n")
        f.write(f"  leak_rate: {row['leak_rate']:.4f}\n")
        f.write(f"  empty_rate: {row['empty_rate']:.4f}\n\n")

    f.write("2. BEST PROMPT BY CATEGORY\n")
    f.write("-" * 100 + "\n")
    for _, row in best_prompt_report_df.iterrows():
        f.write(f"Category: {row['subcategory']}\n")
        if "prompt_variant" in row:
            f.write(f"  prompt_variant: {row['prompt_variant']}\n")
        if "mean_proxy_score" in row:
            f.write(f"  mean_proxy_score: {row['mean_proxy_score']}\n")
        if "selection_score" in row:
            f.write(f"  selection_score: {row['selection_score']}\n")
        f.write("  BEST PROMPT:\n")
        f.write(str(row["prompt_text"]) + "\n\n")

    f.write("3. TOP GENERATED CUES BY CATEGORY\n")
    f.write("-" * 100 + "\n")
    for subcat in sorted(top_cues_by_category["subcategory"].unique()):
        f.write(f"Category: {subcat}\n")
        sub_df = top_cues_by_category[top_cues_by_category["subcategory"] == subcat]
        for _, r in sub_df.iterrows():
            f.write(f"  - target_word: {r['target_word']}\n")
            f.write(f"    reference_best_human_cue: {r['reference_best_human_cue']}\n")
            f.write(f"    final_generated_cue: {r['final_generated_cue']}\n")
            f.write(f"    proxy_score: {r['proxy_score']}\n")
        f.write("\n")

print("Saved:", summary_txt_path)
```
