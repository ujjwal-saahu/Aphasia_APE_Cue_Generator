```python
import sys
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT / "src"))
```

```python
FINAL_DIR = PROJECT_ROOT / "data" / "final"

BEST_PROMPT_PATH = FINAL_DIR / "best_prompt_by_category.csv"
FINAL_CUES_PATH = FINAL_DIR / "final_generated_cues_by_category.csv"
MOBILE_PACKAGE_PATH = FINAL_DIR / "mobile_ready_cue_image_package.csv"

print("BEST_PROMPT_PATH:", BEST_PROMPT_PATH)
print("FINAL_CUES_PATH:", FINAL_CUES_PATH)
print("MOBILE_PACKAGE_PATH:", MOBILE_PACKAGE_PATH)
```

```python
best_prompt_df = pd.read_csv(BEST_PROMPT_PATH)
final_cues_df = pd.read_csv(FINAL_CUES_PATH)
mobile_package_df = pd.read_csv(MOBILE_PACKAGE_PATH)

print("best_prompt_df:", best_prompt_df.shape)
print("final_cues_df:", final_cues_df.shape)
print("mobile_package_df:", mobile_package_df.shape)

display(best_prompt_df.head(10))
display(final_cues_df.head(10))
display(mobile_package_df.head(10))
```

```python
for df in [best_prompt_df, final_cues_df, mobile_package_df]:
    for col in df.columns:
        if df[col].dtype == object:
            df[col] = df[col].fillna("").astype(str).str.strip()
```

```python
best_prompt_keep = [c for c in [
    "subcategory",
    "prompt_variant",
    "mean_proxy_score",
    "selection_score",
    "prompt_text"
] if c in best_prompt_df.columns]

best_prompt_merge_df = best_prompt_df[best_prompt_keep].copy()
best_prompt_merge_df = best_prompt_merge_df.rename(columns={
    "prompt_text": "best_prompt_full",
    "mean_proxy_score": "best_prompt_score"
})

final_demo_df = final_cues_df.merge(
    best_prompt_merge_df,
    on="subcategory",
    how="left"
)

print("After merging best prompts:", final_demo_df.shape)
display(final_demo_df.head(10))
```

```python
final_demo_df = final_demo_df.merge(
    mobile_package_df[["target_word", "subcategory", "image_path"]],
    on=["target_word", "subcategory"],
    how="left"
)

print("After merging image paths:", final_demo_df.shape)
display(final_demo_df.head(10))
```

```python
final_demo_export_df = final_demo_df[[
    "target_word",
    "pos",
    "main_category",
    "subcategory",
    "reference_best_human_cue",
    "final_generated_cue",
    "proxy_score",
    "semantic_similarity",
    "brevity_score",
    "target_leak_penalty",
    "prompt_variant",
    "best_prompt_score",
    "selection_score",
    "best_prompt_full",
    "image_path",
]].copy()

final_demo_export_df = final_demo_export_df.sort_values(
    ["subcategory", "target_word"],
    ascending=[True, True]
).reset_index(drop=True)

print("FINAL DEMO EXPORT SHAPE:", final_demo_export_df.shape)
display(final_demo_export_df.head(20))
```

```python
final_demo_package_path = FINAL_DIR / "final_demo_package.csv"
final_demo_export_df.to_csv(final_demo_package_path, index=False, encoding="utf-8-sig")

print("Saved:", final_demo_package_path)
```

```python
mobile_app_export_df = final_demo_export_df[[
    "target_word",
    "subcategory",
    "final_generated_cue",
    "image_path",
]].copy()

mobile_app_export_path = FINAL_DIR / "mobile_app_export.csv"
mobile_app_export_df.to_csv(mobile_app_export_path, index=False, encoding="utf-8-sig")

print("Saved:", mobile_app_export_path)
display(mobile_app_export_df.head(20))
```

```python
professor_review_df = final_demo_export_df[[
    "target_word",
    "subcategory",
    "reference_best_human_cue",
    "final_generated_cue",
    "proxy_score",
    "best_prompt_full",
    "image_path",
]].copy()

professor_review_path = FINAL_DIR / "professor_review_package.csv"
professor_review_df.to_csv(professor_review_path, index=False, encoding="utf-8-sig")

print("Saved:", professor_review_path)
display(professor_review_df.head(20))
```

```python
summary_txt_path = FINAL_DIR / "final_demo_summary.txt"

with open(summary_txt_path, "w", encoding="utf-8") as f:
    f.write("FINAL DEMO PACKAGE SUMMARY\n")
    f.write("=" * 100 + "\n\n")

    for subcat in sorted(final_demo_export_df["subcategory"].unique()):
        f.write("=" * 100 + "\n")
        f.write(f"CATEGORY: {subcat}\n\n")

        sub_df = final_demo_export_df[final_demo_export_df["subcategory"] == subcat].copy()

        best_prompt_text = sub_df["best_prompt_full"].iloc[0] if len(sub_df) > 0 else ""
        f.write("BEST PROMPT:\n")
        f.write(str(best_prompt_text) + "\n\n")

        f.write("TOP ITEMS:\n")
        top_rows = sub_df.sort_values("proxy_score", ascending=False).head(10)

        for _, r in top_rows.iterrows():
            f.write(f"- target_word: {r['target_word']}\n")
            f.write(f"  reference_best_human_cue: {r['reference_best_human_cue']}\n")
            f.write(f"  final_generated_cue: {r['final_generated_cue']}\n")
            f.write(f"  proxy_score: {r['proxy_score']}\n")
            f.write(f"  image_path: {r['image_path']}\n\n")

print("Saved:", summary_txt_path)
```

```python
print("All final exports created successfully.")
print()
print("1. final_demo_package.csv")
print("2. mobile_app_export.csv")
print("3. professor_review_package.csv")
print("4. final_demo_summary.txt")
```
