```python
import sys
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT / "src"))
```

```python
from config import CATEGORY_INITIAL_RESULTS_PATH, CATEGORY_ROUND0_SUMMARY_PATH
from scoring_utils import compute_proxy_score
```

```python
initial_df = pd.read_csv(CATEGORY_INITIAL_RESULTS_PATH)

print("INITIAL_DF SHAPE:", initial_df.shape)
print("INITIAL_DF COLUMNS:", initial_df.columns.tolist())
display(initial_df.head(10))
```

```python
score_rows = []

for _, row in initial_df.iterrows():
    score_dict = compute_proxy_score(
        generated_cue=row["generated_cue"],
        target_word=row["target_word"],
        reference_cue=row["reference_cue"],
    )

    score_rows.append({
        **row.to_dict(),
        **score_dict,
    })

scored_df = pd.DataFrame(score_rows)

print("SCORED_DF SHAPE:", scored_df.shape)
display(scored_df.head(10))
```

```python
# keep only rows where generated cue is not empty
scored_df = scored_df[
    scored_df["generated_cue"].fillna("").astype(str).str.strip() != ""
].copy()

# fill grouping columns so groupby does not drop everything
group_cols = [
    "ape_round",
    "parent_prompt_id",
    "subcategory",
    "prompt_id",
    "prompt_family",
    "prompt_variant",
    "prompt_text",
]

for col in ["parent_prompt_id", "subcategory", "prompt_id", "prompt_family", "prompt_variant", "prompt_text"]:
    scored_df[col] = scored_df[col].fillna("").astype(str)

scored_df["ape_round"] = scored_df["ape_round"].fillna(0)

print("AFTER CLEANING SHAPE:", scored_df.shape)
display(scored_df.head(10))
```

```python
round0_summary = (
    scored_df.groupby(
        group_cols,
        as_index=False,
        dropna=False,
    )
    .agg(
        mean_proxy_score=("proxy_score", "mean"),
        mean_similarity=("semantic_similarity", "mean"),
        mean_brevity=("brevity_score", "mean"),
        mean_target_leak=("target_leak_penalty", "mean"),
    )
    .sort_values(["subcategory", "mean_proxy_score"], ascending=[True, False])
    .reset_index(drop=True)
)

print("ROUND0 SUMMARY SHAPE:", round0_summary.shape)
print("ROUND0 SUMMARY COLUMNS:", round0_summary.columns.tolist())
display(round0_summary.head(20))
```

```python
round0_summary.to_csv(CATEGORY_ROUND0_SUMMARY_PATH, index=False, encoding="utf-8-sig")
print("Saved:", CATEGORY_ROUND0_SUMMARY_PATH)
```
